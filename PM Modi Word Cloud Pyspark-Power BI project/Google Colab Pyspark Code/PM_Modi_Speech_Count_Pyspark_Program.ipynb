```python

PM_Speech = spark.sparkContext.textFile("/content/drive/MyDrive/spark/PM Modi 76th Independnce Speech.txt")

type(PM_Speech)
```

```python
split_by=PM_Speech.flatMap(lambda x:x.split(" ")) 
```

```python
mapped=split_by.map(lambda x:(x,1))
```

```python
reduce=mapped.reduceByKey(lambda x,y:x+y)
```

```python
sort=reduce.sortByKey()
```

```python
def lower_clean_str(x):
  punc='!"#$%&\'()*+“,-./:;<=‘>”?@[\\]^_`"{|}~'
  lowercased_str = x.lower()
  for ch in punc:
    lowercased_str = lowercased_str.replace(ch, '')
  return lowercased_str

one_RDD = sort.map(lambda x:lower_clean_str(x[0]))
one_RDD.collect()
```

```python
removeStopWords=one_RDD.filter(lambda x: x not in stopwords)
```

```python
removeStopWords.count()
```

```python
removeStopWords.collect()
```

```python
mapped= removeStopWords.map(lambda x:(x,1))
```

```python
reduce=mapped.reduceByKey(lambda x,y:x+y)
```

```python
ordered=reduce.map(lambda x:(x[1],x[0]))
```

```python
final=ordered.sortByKey(False)
final1=final.map(lambda x:x[1])
```

```python
final1.collect()
```

```python
def Upper_case_str(x):

  uppercased_str = x.upper()
  
  return uppercased_str
```

```python
finalUpperCase = final1.map(lambda x:Upper_case_str(x))
finalUpperCase.collect()
```
